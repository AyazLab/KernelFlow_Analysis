```python
df = par.flow.create_source_detector_df("3D", add_missing=True, MNI=True, brain_regions=True)
```

```python
# XYZ coordinates
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.view_init(azim=120, elev=30)
ax.scatter(df["detector_x_pos"], df["detector_y_pos"], df["detector_z_pos"])
ax.scatter(df["source_x_pos"], df["source_y_pos"], df["source_z_pos"])
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("XYZ coordinates")
plt.show()

# MNI coordinates
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.view_init(azim=120, elev=30)
ax.scatter(df["detector_x_MNI"], df["detector_y_MNI"], df["detector_z_MNI"])
ax.scatter(df["source_x_MNI"], df["source_y_MNI"], df["source_z_MNI"])
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("MNI coordinates")
plt.show()
```
